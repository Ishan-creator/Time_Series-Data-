```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt 
import datetime as datetime 
%matplotlib inline 
```

```python
df = yf.download('TSLA')
```

```python
df.head()
```

```python
df.tail()
```

```python
df['Open'].plot(figsize = (15,6))
```

```python
df['Rolling'] = df['Open'].rolling(10, min_periods=1).mean()
```

```python
df.head(11)
```

```python
df[['Open','Rolling']].plot(xlim = ['2022-01-01', '2024-01-01'] ,figsize = (20,8))
```

```python
df['Rolling_30']= df["Open"].rolling(30).mean()
```

```python
df['Rolling_50']= df["Open"].rolling(50).mean()
```

```python
df[['Open','Rolling','Rolling_30','Rolling_50']].plot(xlim = ['2022-01-01', '2024-01-01'] ,figsize = (20,8))
```

```python
df['EMA'] = df['Open'].ewm(alpha = 0.1 , adjust = False).mean()
```

```python
df[['EMA' , 'Open']].plot(xlim = ['2021-01-01' , '2023-01-01'],figsize = (15,6))
```

```python
df['EWMA'] = df['Open'].ewm(span = 5).mean()
```

```python
df[['Open', 'EWMA']].plot(figsize = (15,6))
```
